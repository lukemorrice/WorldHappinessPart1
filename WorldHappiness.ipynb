# World Happiness - Part 1

## Modelling the Correlation Between Country Happiness and Internet Usage

This notebook inspects the correlation between happiness per country and the internet usage per country (for populations of at least 1 million). I'm curious to inspect whether the level of internet usage has any correlation/impact on the level of happiness in a country. More specifically, I want to determine how social media usage affects a country's percieved happiness by its residents. I will explore this aspect in part 2. In this notebook, I will try and discover if there is indeed any correlation.

Firstly, we need to import Pandas library and the datasets.

```python
import pandas as pd
```

```python
internet_use1 = pd.read_csv('internet_use.csv')
```

```python
df1 = internet_use1
df1.head()
```

Dropping unnecessary columns from the dataset.

```python
df1.drop(['Rank', 'Percentage', 'Rank.1'], axis=1, inplace=True)
```

### Manipulation
Converting the column data types from strings to integers using the string_to_int function I've defined below. We apply this to the columns 'Internet Users' and 'Population'. Then, I create a new column to normalise the internet users according to the country's population size.

```python
def string_to_int(x):
    x = int(x.replace(',', ''))
    return float(x)
```

```python
df1['Internet Users'] = df1['Internet Users'].apply(string_to_int)
```

```python
df1['Population'] = df1['Population'].apply(string_to_int)
```

```python
df1.drop(df1[df1['Population'] < 1000000].index, inplace=True)
```

```python
df1['Users Per Population'] = round(df1['Internet Users'] / df1['Population'] * 100, 2)
```

Lastly for the manipulation side of things, I sort the normalised internet useage in descending order and then reset the dataframe's index.

```python
df1.sort_values('Users Per Population', ascending=False, inplace=True)
```

```python
df1 = df1.reset_index()
df1.drop(['index'], axis=1, inplace=True)
```

```python
df1
```

## Graphing the Data
Creating some basic graphs to visualise the new dataset. Will use Seaborn to do this.

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
df1.describe()
```

```python
df1
```

```python
top_10 = df1[:10]
fig, ax = plt.subplots(figsize=(9, 6))
sns.barplot(ax=ax, x=top_10.index, y="Users Per Population", data=top_10)
ax.set(ylim=(90, 100))
ax.set_ylabel('% of Internet Users per Population', fontsize=16)
ax.set_xticklabels(labels=top_10['Country'], rotation=45, fontsize=14)
ax.tick_params(axis='y', labelsize=14)
ax.set_title('(Normalised) Internet Users per Country', fontsize=18)
plt.show()
```

```python
df1.rename(columns = {'Country or Area' : 'Country'}, inplace=True)
```

## World Happiness Dataset

```python
df2 = pd.read_csv('happiness.csv')
df2.head()
df2.drop(df2.iloc[:, 3:], inplace=True, axis=1)
```

#### Join Dataframes on Country Column

```python
df3 = df1.merge(df2, how='inner', on=['Country'])
df3.sort_values('Happiness.Rank', ascending=True, inplace=True)
df3
```

## Graphing the Data, Again
Graphing this new dataset, now ordered by the happiness rank of country. Included a horizontal line to visualise the mean internet usage per country. To keep the numbers somewhat consistent and comparable, I will limit the subset of countries to only be those in Europe.

```python
continents = pd.read_csv('continents.csv', encoding='ISO-8859-1')
continents = continents.loc[:, ['country', 'continent']]
continents = continents[continents['continent'] == 'Europe']
continents.rename(columns={'country':'Country', 'continent':'Continent'}, inplace=True)
continents['Country'].iloc[-1:] = 'United Kingdom'
```

```python
df3 = df3.merge(continents, how='inner', on=['Country'])
df3
```

```python
mean_internet_usage = df3['Users Per Population'].mean()
mean_internet_usage
```

Shortening the label of Bosnia and Herzegovina to clean up the x-tick labels on the graph.

```python
df3['Country'].iloc[28] = 'Bosnia'
```

```python
fig, ax = plt.subplots(figsize=(20, 6))
sns.barplot(ax=ax, x=df3.index, y="Users Per Population", data=df3)
ax.axhline(mean_internet_usage, ls='--')
ax.set(ylim=(55, 100))
ax.set_ylabel('% of Internet Users per Population', fontsize=16)
ax.set_xticklabels(labels=df3['Country'], rotation=45, fontsize=14)
ax.tick_params(axis='y', labelsize=14)
ax.set_title('Normalised Internet Users per Country', fontsize=18)
plt.show()
```

### Conclusion
As we can see from the above graph, there is no obvious correlation between internet usage and a country's happiness. If anything, there appears to be a downwards slope, indicating that happier countries correlate with a higher access to the internet. However, this graph does not consider ease of access to the internet and hence we cannot conclude anything with any certainty.
